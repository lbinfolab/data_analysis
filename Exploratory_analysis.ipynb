```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
data = pd.read_csv("seattle-weather.csv", index_col='date', parse_dates=True)
data
```

```python
plt.rcParams["figure.figsize"] = (20,6)
x = data.index
plt.plot(x, data.precipitation)
plt.plot(x, data.temp_max)
plt.plot(x, data.wind)

plt.show()
```

```python
data.corr(method='pearson')
```

```python
plt.rcParams["figure.figsize"] = (5,5)

pd.plotting.lag_plot(data['precipitation'], lag=1)
plt.show()
pd.plotting.lag_plot(data['temp_max'], lag=1)
plt.show()
pd.plotting.lag_plot(data['wind'], lag=1)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12,3)
plot_acf(data['temp_max'])
plot_acf(data['precipitation'])
plot_acf(data['wind'])
plt.show()
```

```python
plot_pacf(data['temp_max'])
plot_pacf(data['precipitation'])
plot_pacf(data['wind'])
plt.show()
```

```python
pd.plotting.autocorrelation_plot(data["temp_max"])
plt.show()
pd.plotting.autocorrelation_plot(data["precipitation"])
plt.show()
pd.plotting.autocorrelation_plot(data["wind"])
plt.show()
```

```python
data['wind'].plot(kind='kde')
plt.show()
data['temp_max'].plot(kind='kde')
plt.show()
data['precipitation'].plot(kind='hist', bins=30)
plt.show()
```
